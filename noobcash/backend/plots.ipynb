```python
import numpy as np
import matplotlib.pyplot as plt

N = 3
ind = np.arange(N)  # the x locations for the groups
width = 0.27       # the width of the bars

fig = plt.figure()
ax = fig.add_subplot(111)

yvals = [4, 9, 2]
rects1 = ax.bar(ind, yvals, width, color='g')
zvals = [1,2,3]
rects2 = ax.bar(ind+width, zvals, width, color='m')
plt.title('Capacity')
ax.set_ylabel('Capacity')
plt.xlabel('experiment setting')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('c=1,d=4', 'c=1,d=5', 'c=5,d=5','c=10,d=4','c=10,d=5') )
ax.legend( (rects1[0], rects2[0]), ('5 clients', '10 clients') )
plt.grid()
def autolabel(rects):
    for rect in rects:
        h = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*h, '%d'%int(h),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
```
